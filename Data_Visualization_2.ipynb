### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing dataset

```python
df = pd.read_csv(r"Regiment.txt",sep='\t')
df
```

```python
df.plot.bar()
```

### Line Plot

```python
plt.figure(figsize=(12,5))
plt.plot(df.index,df['postTestScore'],c='r',alpha=0.2)
plt.scatter(df.index,df['postTestScore'],c='g')
plt.xticks(np.arange(0,12),labels=df['name'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.index,df['preTestScore'],c='r',alpha=0.2)
plt.scatter(df.index,df['preTestScore'],c='g')
plt.xticks(np.arange(0,12),labels=df['name'])
plt.show()
```

### Pie Chart

```python
reg = df.groupby(['regiment'])['postTestScore'].mean()
plt.pie(reg,labels=reg.index,autopct="%d")   # useof autopct
plt.show()
```

## Catplot
---

```python
sns.catplot(x="regiment", y="preTestScore",hue="company", data=df)
```

```python
sns.catplot(x="regiment", y="postTestScore",hue="company",jitter=False, data=df)
```
